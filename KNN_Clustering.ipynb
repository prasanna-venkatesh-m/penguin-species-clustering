```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/penguins.csv')
```

```python
print(df.isnull().sum())
```

```python
drop_df = df.dropna()
```

```python
drop_df.isnull().sum()
```

```python
drop_df.head()
```

```python
print(drop_df.isnull().sum())
```

```python
drop_df[drop_df["flipper_length_mm"] > 4000]
drop_df[drop_df["flipper_length_mm"] < 0]
```

```python
cleaned_data = drop_df.drop(drop_df[drop_df["flipper_length_mm"] > 4000].index)
cleaned_data = cleaned_data.drop(cleaned_data[cleaned_data["flipper_length_mm"] < 0].index)
```

```python
cleaned_data.isnull().sum()
```

```python
cleaned_data.head()
```

```python
ohe = OneHotEncoder()
o = ohe.fit_transform(cleaned_data[['sex']])
ohe_df = pd.DataFrame(o.toarray(), columns=ohe.get_feature_names_out(['sex']))
ohe_df.head()
```

```python
ohe_df.isnull().sum()
```

```python
ohe_data = pd.concat([cleaned_data.reset_index(drop=True), ohe_df.reset_index(drop=True)], axis=1)
```

```python
ohe_data.isnull().sum()
```

```python
ohe_data.head()
```

```python
removed_data = ohe_data.drop(columns=['sex','sex_.'])
```

```python
removed_data.head()
```

```python
sc = StandardScaler()
```

```python
scaled_data = sc.fit_transform(removed_data)
```

```python
scaled_data
```

```python
data = pd.DataFrame(scaled_data, columns=removed_data.columns)
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data)
```

```python
new_data = pd.DataFrame(pca_data)
new_data
```

```python
plt.figure(figsize=(8,6))
plt.scatter(new_data[0], new_data[1], c='r')
plt.show()
```

```python
inertia = []
clusters = []

for i in range(1, 10):
  knn = KMeans(n_clusters=i, random_state=42)
  knn.fit_transform(new_data)
  inertia.append(knn.inertia_)
  clusters.append(i)
```

```python
plt.figure(figsize=(8,6))
plt.plot(clusters, inertia, 'bx-')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
knn = KMeans(n_clusters=4)
knn.fit(new_data)
```

```python
print(knn.labels_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(new_data[0], new_data[1], c=knn.labels_,  cmap="viridis")
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
print(knn.predict([[1,2]]))
```
